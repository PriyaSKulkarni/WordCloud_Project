Mini project 1: Text Summary

```python
text= """The book is structured around the metaphor of a staircase, where each step represents a critical element of reaching personal and professional success. The first step, according to Ziglar, is Self-Image. He emphasizes the importance of having a positive self-image and how it is foundational to achieving anything meaningful in life.

The next step is building good Character, followed by the significance of having the right Attitude. Ziglar suggests that a positive attitude enables you to face challenges head-on and is instrumental in achieving success.

Work and effort are also major themes in the book. Ziglar emphasizes that setting Goals is another critical step on the staircase. Without goals, according to him, it's easy to drift aimlessly through life. He encourages the reader to write down their short-term and long-term goals and devise action plans to achieve them.

Interpersonal relationships and creating a healthy environment by being around positive, uplifting people is another point that Ziglar covers. The book delves into various aspects of human psychology and social interactions that can either elevate you or bring you down.

Lastly, Ziglar discusses Desire, the final step on the staircase. He argues that without a burning desire to achieve your goals, all the other steps are inconsequential. Ziglar doesn't just present abstract ideas; he provides practical advice and actionable steps one can take to start climbing their personal staircase to the top."""
```

```python
# Count the frequency of each word from corpus
# Deal with stopwords
# visualize with wordcloud
```


Count the frequency

```python
from re import I
words={}
for i in text.split():
  i = i.lower().strip()
  if i in words:
    words[i] +=1
  else:
    words[i] =1
print(words)
```

```python
# total number od values
count = sum(words.values())
print(count)
```

Stopwords

```python
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
```

```python
#nltk.download('stopwords')
stop_word = set(stopwords.words('english'))
print(f"number of stopwords: {len(stop_word)}")
print(stop_word)
```

```python
stop_word.update(["book",'ziglar'])
print(stop_word)
print(f"number of stopwords: {len(stop_word)}")
```

```python
filtered_text = " ".join([word for word in text.split() if word.lower() not in stop_word])
filtered_text
```

```python
import string
st = ""
for i in filtered_text:
  if i not in string.punctuation:
    st+=i
print(st)
```

```python
filtered_text = " ".join([word for word in st.split() if word.lower() not in stop_word])
filtered_text
```

```python
from re import I
words={}
for i in filtered_text.split():
  i = i.lower().strip()
  if i in words:
    words[i] +=1
  else:
    words[i] =1
print(words)
```

```python
count = sum(words.values())
print(count)
```
